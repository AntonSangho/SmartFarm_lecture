### **Data 요약 정보 설명**

| **항목**         | **설명**                                                |
|------------------|---------------------------------------------------------|
| **count**        | 해당 열(변수)의 **데이터 개수**를 나타냅니다. NaN(결측치)은 제외됩니다. |
| **mean**         | **평균값**을 나타냅니다. (모든 값을 더한 후 데이터 개수로 나눈 값) |
| **std**          | **표준편차**를 나타냅니다. (데이터의 분산 정도를 나타내는 지표) |
| **min**          | 해당 열의 **최소값**을 보여줍니다.                      |
| **25%**          | **1사분위수(25th Percentile)**: 전체 데이터의 하위 25%에 해당하는 값 |
| **50%** (median) | **중앙값(50th Percentile)**: 데이터의 가운데 위치한 값입니다. |
| **75%**          | **3사분위수(75th Percentile)**: 전체 데이터의 상위 25%에 해당하는 값 |
| **max**          | 해당 열의 **최대값**을 보여줍니다.                      |


```python
# pandas와 scipy 라이브러리 불러오기
import pandas as pd
from scipy import stats

# CSV 파일 경로
path_AILab = "./data-A.csv"
path_Multimedia = "./data-M.csv"

# 두 CSV 파일을 DataFrame으로 읽어오기
data_AILab = pd.read_csv(path_AILab, names=['DateTime', 'Humidity', 'Temperature', 'aqi', 'eco2', 'tvoc'])
data_Multimedia = pd.read_csv(path_Multimedia, names=['DateTime', 'Humidity', 'Temperature', 'aqi', 'eco2', 'tvoc'])

# 데이터 요약 통계 출력
print("AILab 데이터 요약 통계:")
print(data_AILab.describe())

print("\nMultimedia 데이터 요약 통계:")
print(data_Multimedia.describe())
```

# 2.비교하기
## 2-1. 온도

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
path_AILab = "./data-A.csv"
path_Multimedia = "./data-M.csv"

# 데이터 로드
data_AILab = pd.read_csv(path_AILab)
data_Multimedia = pd.read_csv(path_Multimedia)

# 열 이름 설정
data_AILab.columns = ['DateTime', 'Humidity', 'Temperature', 'aqi', 'eco2', 'tvoc']
data_Multimedia.columns = ['DateTime', 'Humidity', 'Temperature', 'aqi', 'eco2', 'tvoc']

# 시간 형식 변환
data_AILab['DateTime'] = pd.to_datetime(data_AILab['DateTime'])
data_Multimedia['DateTime'] = pd.to_datetime(data_Multimedia['DateTime'])

# 시각화를 위한 선 그래프 생성
plt.figure(figsize=(12, 6))

# AILab 데이터의 온도 변화
plt.plot(data_AILab['DateTime'], data_AILab['Temperature'], label='AILab Temperature', linewidth=2)

# Multimedia 데이터의 온도 변화
plt.plot(data_Multimedia['DateTime'], data_Multimedia['Temperature'], label='Multimedia Temperature', linewidth=2)

# 차트 제목과 축 레이블 설정
plt.title('Temperature Comparison Over Time (AILab vs Multimedia)', fontsize=20)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Temperature (°C)', fontsize=14)

# x축 라벨 회전 및 레이아웃 조정
plt.xticks(rotation=45)
plt.legend()  # 범례 표시
plt.tight_layout()

# 그래프 출력
plt.show()

```

## 2-3. AQI

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
path_AILab = "./data-A.csv"
path_Multimedia = "./data-M.csv"

# 데이터 로드
data_AILab = pd.read_csv(path_AILab)
data_Multimedia = pd.read_csv(path_Multimedia)

# 열 이름 설정
data_AILab.columns = ['DateTime', 'Humidity', 'Temperature', 'aqi', 'eco2', 'tvoc']
data_Multimedia.columns = ['DateTime', 'Humidity', 'Temperature', 'aqi', 'eco2', 'tvoc']

# 시간 형식 변환
data_AILab['DateTime'] = pd.to_datetime(data_AILab['DateTime'])
data_Multimedia['DateTime'] = pd.to_datetime(data_Multimedia['DateTime'])

# 날짜별 평균 AQI 계산
data_AILab['Date'] = data_AILab['DateTime'].dt.date
data_Multimedia['Date'] = data_Multimedia['DateTime'].dt.date

avg_aqi_AILab = data_AILab.groupby('Date')['aqi'].mean().reset_index()
avg_aqi_Multimedia = data_Multimedia.groupby('Date')['aqi'].mean().reset_index()

# 막대 그래프 생성
fig, ax = plt.subplots(figsize=(12, 6))

# AILab 데이터의 막대 그래프
ax.bar(avg_aqi_AILab['Date'], avg_aqi_AILab['aqi'], width=0.4, label='AILab AQI', align='center')

# Multimedia 데이터의 막대 그래프 (x축 위치를 조정)
ax.bar(avg_aqi_Multimedia['Date'], avg_aqi_Multimedia['aqi'], width=0.4, label='Multimedia AQI', align='edge')

# 차트 제목과 축 레이블 설정
ax.set_title('Average AQI Comparison (AILab vs Multimedia)', fontsize=20)
ax.set_xlabel('Date', fontsize=14)
ax.set_ylabel('AQI', fontsize=14)

# x축 라벨 회전 및 범례 추가
plt.xticks(rotation=45)
ax.legend()

# 레이아웃 조정 및 그래프 출력
plt.tight_layout()
plt.show()
```
