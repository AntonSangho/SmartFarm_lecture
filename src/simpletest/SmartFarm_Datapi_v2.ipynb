```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
csv_file = "./data.csv"

# CSV 파일을 DataFrame으로 읽어오기
data = pd.read_csv(csv_file, names=['DateTime', 'Humidity', 'Temperature', 'aqi', 'eco2', 'tvoc'])

# 데이터 확인
print(data.head())

# 시각화 설정
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()

# scatter plot 생성
scatter = ax.scatter(data['DateTime'], data['Temperature'], cmap=plt.cm.Blues, s=50)

# 차트 제목과 축 라벨 설정
ax.set_title("Temperature Over Time", fontsize=24)
ax.set_xlabel("Time", fontsize=14)
ax.set_ylabel("Temperature (°C)", fontsize=14)

# x 축 라벨이 겹치지 않도록 설정
plt.xticks(rotation=45)

plt.tight_layout()  # 그래프가 잘리는 것을 방지하기 위해 레이아웃 조정
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 경로
csv_file = "./data.csv"

# CSV 파일을 DataFrame으로 읽어오기
data = pd.read_csv(csv_file, names=['DateTime', 'Humidity', 'Temperature', 'aqi', 'eco2', 'tvoc'])

# 시각화 설정
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()

# scatter plot 생성
scatter = ax.scatter(data['DateTime'], data['Temperature'], cmap=plt.cm.Blues, s=50)

# 차트 제목과 축 라벨 설정
ax.set_title("Temperature", fontsize=24)
ax.set_xlabel("Date and Time", fontsize=14)
ax.set_ylabel("Temperature (°C)", fontsize=14)

# x 축 라벨이 겹치지 않도록 설정
plt.xticks(rotation=45)
ax.xaxis.set_major_locator(plt.MaxNLocator(10))  # x 축 라벨 간격 설정

plt.tight_layout()  # 그래프가 잘리는 것을 방지하기 위해 레이아웃 조정
plt.show()
```
